# ***SQUID GAME SENTIMENT ANALYSIS***

This Python project aims for sentiment analysis to understand people's future reactions to Squid Game.

```python
# Importing essential libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

The Dataset for the given project is taken from kaggle dataset.


```python
#Importing the dataset taken from kaggle
data = pd.read_csv('/content/Project-1.csv')
data.head()
```

**Data Pre-processing and cleaning of data **

```python
# Dropping user_location column
data = data.drop(columns="user_location", axis=1)
```

```python
#Checking the dataset values and counting if any null values
print(data.isnull().sum())
```

```python
#Analysing the values in the dataset, we can clearly identify user_description have 5211 null values.Hence dropping the column.
data = data.drop(columns = "user_description" , axis = 1)
data = data.dropna()
```

Importing nltk (Natural language toolkit) and re(Regular expression)

```python
import nltk
import re
nltk.download('stopwords')
stemmer = nltk.SnowballStemmer("english")
from nltk.corpus import stopwords
import string
stopword=set(stopwords.words('english'))

def clean(text):
    text = str(text).lower()
    text = re.sub('\[.*?\]', '', text)
    text = re.sub('https?://\S+|www\.\S+', '', text)
    text = re.sub('<.*?>+', '', text)
    text = re.sub('[%s]' % re.escape(string.punctuation), '', text)
    text = re.sub('\n', '', text)
    text = re.sub('\w*\d\w*', '', text)
    text = [word for word in text.split(' ') if word not in stopword]
    text=" ".join(text)
    text = [stemmer.stem(word) for word in text.split(' ')]
    text=" ".join(text)
    return text
data["text"] = data["text"].apply(clean)
```

```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud , STOPWORDS

text = " ".join(i for i in data.text)
stopwords = set(STOPWORDS)
wordcloud = WordCloud(stopwords=stopwords, background_color="black").generate(text)
plt.figure( figsize=(15,10))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer

nltk.download('vader_lexicon')
sentiments = SentimentIntensityAnalyzer()
data["Positive"] = [sentiments.polarity_scores(i)["pos"] for i in data["text"]]
data["Negative"] = [sentiments.polarity_scores(i)["neg"] for i in data["text"]]
data["Neutral"] = [sentiments.polarity_scores(i)["neu"] for i in data["text"]]
data = data[["text", "Positive", "Negative", "Neutral"]]
data.head()
```

```python
x = sum(data["Positive"])
y = sum(data["Negative"])
z = sum(data["Neutral"])

def sentiment_score(a, b, c):
    if (a>b) and (a>c):
        print("Positive üòä ")
    elif (b>a) and (b>c):
        print("Negative üò† ")
    else:
        print("Neutral üôÇ ")
sentiment_score(x, y, z)
```
